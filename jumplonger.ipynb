```python
import math
class Jumper():
    high = 0
    distance = 0
    velocity = 0
    angle = 0
    acc = -10
    def __init__(self,V,Ag,Acc=-9.8,D=0,H=0):
        self.high = H
        self.distance = D
        self.velocity = V
        angle = Ag/180*math.pi
        self.acc = Acc
        self.v_v0 = V*math.sin(angle)
        self.v_h0 = V*math.cos(angle)
    def get_position(self,t):
        high1 = self.high + self.v_v0*t + (1/2)*self.acc*t**2
        distance1 = self.distance + self.v_h0*t
        return(distance1, high1)
    def fly_time(self):
        time = -self.v_v0/self.acc*2
        return(time)
    def max_distance(self):
        max_dis = self.distance + self.v_h0*self.fly_time()
        return(max_dis)
```

```python
zzt = Jumper(5,45)
print("fly time = ",zzt.fly_time())
print("Distance-max = ",zzt.max_distance())
h = []
d = []
n=50
for i in range(n+1):
    t = i/n*zzt.fly_time()
    d.append(zzt.get_position(t)[0])
    h.append(zzt.get_position(t)[1])
import matplotlib.pyplot as plt
plt.plot(d,h)
plt.xlim(0,3)
plt.ylim(0,0.7)
# plt.rcParams['font.sans-serif'] = ['SimHei']
# plt.title("Jumping trajectory")
# plt.xlabel("跳跃距离")
plt.show()
```

```python
for i in range(52):
    plt.plot(d[:i],h[:i])
    plt.xlim(0,3)
    plt.ylim(0,0.7)
    plt.savefig("fig/jump/{}.png".format(i))
    plt.cla()
```

```python
def angle_dist(ang_s, ang_e):
    dis_max = []
    angs = []
    for i in range(101):
        an = i/100*(ang_e - ang_s) + ang_s
        bob = Jumper(5,an)
        angs.append(an)
        dis_max.append(bob.max_distance())
    plt.plot(angs, dis_max)
    plt.show()

ang_s = 0
ang_e = 90
angle_dist(ang_s, ang_e)
```
